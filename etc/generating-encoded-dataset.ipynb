### Trace Acquisition

- Program the Arduino device to run a specific loop program.
- Run the GRC script so that it keeps publishing I-Q data samples of that activity.
- In this Jupyter Notebook in order to capture EM traces into Numpy files and then generate the statistical metrics dataset of the EM traces.

### Importing libraries

```python
import numpy as np
import scipy.stats as scpstat
import time
from emvincelib import iq, ml, stat
from scipy.fftpack import fft
from scipy import signal
%matplotlib inline
import matplotlib.pyplot as plt
```

### Encoded Signal Generation

- Take a channel
- Move a non-overlapping sliding window through the channel
- Calculate the statistical properties for every window and store in an array
- Collect the set of array containing statistical properties and append them into a single 2D array
- Finally append all the statistical property arrays in to a single matrix

### Configurations

```python
# size of the Fourer Transform window in seconds.
windowStepSize = 1e-3 # 1 millisecond

# length of each I-Q trace
sampleDuration = 500 # 500 milliseconds

# size of a segment taken from the .cfile in each iteration in segments
#iq_segment_size = 2

# sample rate
sampleRate = 20e6

# number of statistics to calculate
numStats = 20

# path to the data directory
directoryPath = "./data/"
```

### Functions

```python
def calcSTFT(npyFileName):
    '''
    This function takes the name of a npy file and calculates the STFT.
    Since the STFT data are too large, they are saved to smaller segment files.
    '''
    
    print(npyFileName)
    path1, path2, filename = npyFileName.split("/")
    print(path1)
    print(path2)
    print(filename)
    
    # checking the time duration of the I-Q data file
    duration = iq.getTimeDuration(npyFileName, fileType="npy")
    print(duration)

    # number of samples per Fourier Transform window
    samples_per_window = int(windowStepSize * sampleRate)
    print(samples_per_window)

    #tempFileName = npyFileName + ".stft"
    tempFileName = path1 + "/" + path2 + "/" + "temp/" + filename + ".stft"
    #data = iq.getSegmentData(cFileName, i, iq_segment_size)
    data = np.load(npyFileName, mmap_mode='r')
    #print(len(data))
    f, t, Zxx = signal.stft(data, fs=sampleRate, nperseg=samples_per_window, noverlap=0)
    print(len(t))
    print(t)
    del data
    np.savez(tempFileName, f=f, t=t, Zxx=Zxx)
    del f, t, Zxx
```
